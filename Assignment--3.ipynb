```python
# Scrape the all billionaires from forbes
driver.get("https://www.forbes.com/billionaires/") 
time.sleep(5)
```

```python
#creating the empty list
rank=[]
name=[]
net_worth=[]
age=[]
citizenship=[]
source=[]
industry=[]

# feteching top 10 billionaire details
ranks = driver.find_elements_by_xpath("//div[@class='rank']")
rank=[]
for i in ranks[0:10]:
    rank.append(i.text)
      
names = driver.find_elements_by_xpath("//div[@class='personName']")
name=[]
for i in names[0:10] :
    name.append(i.text)


net = driver.find_elements_by_xpath("//div[@class='netWorth']")
net_worth=[]
for i in net[0:10] :
    net_worth.append(i.text)
    
age_group = driver.find_elements_by_xpath("//div[@class='age']")
age=[]
for i in age_group[0:10] :
    age.append(i.text)
    
citizenships = driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
citizenship=[]
for i in citizenships[0:10] :
    citizenship.append(i.text)
    
sources = driver.find_elements_by_xpath("//div[@class='source-column']")
source=[]
for i in sources[0:10] :
    source.append(i.text)   
    
industries = driver.find_elements_by_xpath("//div[@class='category']")
industry=[]
for i in industries[0:10] :
    industry.append(i.text)   
```

```python
url = "https://www.google.com/maps/search/indore+longitude+and+latitude/@22.6698364,75.6298758,11.93z";
WebRequest request = WebRequest.Create(url);

using (WebResponse response = (HttpWebResponse)request.GetResponse())
{
    using (StreamReader reader = new StreamReader(response.GetResponseStream(), Encoding.UTF8))
    {
        DataSet dsResult = new DataSet();
        dsResult.ReadXml(reader);
        DataTable dtCoordinates = new DataTable();
        dtCoordinates.Columns.AddRange(new DataColumn[4] { new DataColumn("Id", typeof(int)),
                    new DataColumn("Address", typeof(string)),
                    new DataColumn("Latitude",typeof(string)),
                    new DataColumn("Longitude",typeof(string)) });
        foreach (DataRow row in dsResult.Tables["result"].Rows)
        {
            string geometry_id = dsResult.Tables["geometry"].Select("result_id = " + row["result_id"].ToString())[0]["geometry_id"].ToString();
            DataRow location = dsResult.Tables["location"].Select("geometry_id = " + geometry_id)[0];
            dtCoordinates.Rows.Add(row["result_id"], row["formatted_address"], location["lat"], location["lng"]);
        }
    }
    return dtCoordinates;
}
```
